```python
from sklearn.model_selection import train_test_split 

x = df.drop("Price",axis=1)
y = df["Price"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
model.fit(x_train , y_train)
print("model trained sucessfully")
```

```python
predictions = model.predict(x_test)
print(predictions[:5])
print(y_test[:5])
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
from sklearn.metrics import (
   mean_absolute_error,
   mean_squared_error,
   r2_score
)

import numpy as np

mae = mean_absolute_error(
    y_test,
    predictions
)

mse = mean_squared_error(
    y_test,
    predictions
)

rmse = np.sqrt(mse)

r2 = r2_score(
    y_test,
    predictions
)

print("MAE : ", mae)
print("MSE : ", mse)
print("RMSE : ", rmse)
print("R2 : ", r2)
```

```python
for features , coef in zip (x.columns , model.coef_):
    print(f"{features:15} : {coef:.5f}")
```

```python
import pandas as pd

coef_df = pd.DataFrame({
    "Feature": x.columns,
    "Coefficient": model.coef_
})

coef_df
```

```python
coef_df.sort_values(
    by="Coefficient",
    ascending=False
)
```

```python
coef_df["Importance"] =abs(coef_df["Coefficient"])
```

```python
coef_df.sort_values(
    by="Importance",
    ascending=False
)
```

```python
import matplotlib.pyplot as plt

coef_df.sort_values(
    by="Importance"
).plot(
    x="Feature",
    y="Importance",
    kind="barh",
    legend=False
)

plt.title("Feature Importance")
plt.show()
```
